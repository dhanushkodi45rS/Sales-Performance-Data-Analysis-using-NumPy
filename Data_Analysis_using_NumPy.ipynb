**Sales Performance Data Analysis Using NumPy**

```python
import numpy as np
np.random.seed(42)
sales_data = np.random.randint(20, 100, size = (7, 5))

print("Sales Data(product X Days):\n", sales_data)
print("\nShape of the Sales: ", sales_data.shape)
print("\nSize of the Sales: ", sales_data.size)
print("\nDataType of the Sales: ", sales_data.dtype)

total_sales = np.sum(sales_data)
average_sales = np.mean(sales_data)

print("\n\nTotal Sales: ", total_sales)
print("\nAverage Sales: ", average_sales)

product_sales = np.sum(sales_data, axis = 0)
print("\nTotal Sales of the Product Each Day: ", product_sales)

day_sales = np.sum(sales_data, axis = 1)
print("\nSales By Eachday: ", day_sales)

top_index = np.argmax(product_sales)
print("\nTop Product: ", top_index + 1)
print("\nTotal Sales of the Top Product: ", product_sales[top_index])

high_sales_day = day_sales[day_sales > 300]
print("\nDays with over 300 sales: ", high_sales_day)

avg_sales_day = np.mean(sales_data, axis = 1)
print("\nThe average sales of each day: ", np.round(avg_sales_day, 2))
```
